```python
# Student Data Organizer

students = [] 
subjects_offered = set()

print("Welcome to Student Data Organizer")

while True:
    print("\n1. Add Student\n2. Display All Students\n3. Update Student\n4. Delete Student\n5. Display Subjects\n6. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Name: ")
        age = int(input("Age: "))
        grade = input("Grade: ")
        subs = input("Subjects (comma separated): ").split(",")
        sid = input("Student ID: ")
        dob = input("Date of Birth: ")

        subjects_offered.update(subs)
        student = {
            "id_dob": (sid, dob),
            "name": name,
            "age": age,
            "grade": grade,
            "subjects": subs
        }
        students.append(student)

    elif choice == "2":
        for s in students:
            sid, dob = s["id_dob"]
            print(f"ID: {sid}, Name: {s['name']}, Age: {s['age']}, Grade: {s['grade']}, Subjects: {', '.join(s['subjects'])}")

    elif choice == "3":
        sid = input("Enter Student ID to update: ")
        for s in students:
            if s["id_dob"][0] == sid:
                s["age"] = int(input("New Age: "))
                new_subs = input("New Subjects (comma separated): ").split(",")
                s["subjects"] = new_subs
                subjects_offered.update(new_subs)
                break

    elif choice == "4":
        sid = input("Enter Student ID to delete: ")
        for i, s in enumerate(students):
            if s["id_dob"][0] == sid:
                del students[i]
                break

    elif choice == "5":
        print("Subjects Offered:", ", ".join(subjects_offered))

    elif choice == "6":
        print("Thank you for using Student Data Organizer")
        break
    else:
        print("Invalid choice.")
```
