```python
records = []

class Person:
    def __init__(self, name, age):
        self.name, self.age = name, age

    def details(self):
        return f"Name: {self.name}\nAge: {self.age}"


class Employee(Person):
    def __init__(self, name, age, emp_id, salary):
        super().__init__(name, age)
        self.emp_id, self.salary = emp_id, salary

    def details(self):
        base = super().details()
        return f"{base}\nEmployee ID: {self.emp_id}\nSalary: rs{self.salary}"


class Manager(Employee):
    def __init__(self, name, age, emp_id, salary, dept):
        super().__init__(name, age, emp_id, salary)
        self.dept = dept

    def details(self):
        base = super().details()
        return f"{base}\nDepartment: {self.dept}"


def add_person():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    records.append(Person(name, age))
    print("Person added!")


def add_employee():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    emp_id = input("Enter ID: ")
    salary = float(input("Enter salary: "))
    records.append(Employee(name, age, emp_id, salary))
    print("Employee added!")


def add_manager():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    emp_id = input("Enter ID: ")
    salary = float(input("Enter salary: "))
    dept = input("Enter department: ")
    records.append(Manager(name, age, emp_id, salary, dept))
    print("Manager added!")


def show_all():
    if not records:
        print("No records yet.")
    else:
        for i, r in enumerate(records, 1):
            print(f"\n[{i}]\n{r.details()}")


def compare_salaries():
    employees = [r for r in records if isinstance(r, Employee)]
    if len(employees) < 2:
        print("Not enough employees for comparison.")
        return

    print("\n--- Salary Comparison ---")
    for i in range(len(employees)):
        for j in range(i + 1, len(employees)):
            e1, e2 = employees[i], employees[j]
            print(f"\n{e1.name} (rs{e1.salary}) vs {e2.name} (rs{e2.salary})")
            if e1.salary > e2.salary:
                print(f"→ {e1.name} earns more")
            elif e1.salary < e2.salary:
                print(f"→ {e2.name} earns more")
            else:
                print("→ Equal salary")


while True:
    print("\n--- Employee Management ---")
    print("1. Add Person\n2. Add Employee\n3. Add Manager\n4. Show All\n5. Compare Salaries\n6. Exit")

    ch = input("Choose: ")
    if ch == '1': add_person()
    elif ch == '2': add_employee()
    elif ch == '3': add_manager()
    elif ch == '4': show_all()
    elif ch == '5': compare_salaries()
    elif ch == '6':
        print("Exiting... Bye!")
        break
    else:
        print("Invalid choice")
```
